```python
data['Total_Amount'] = data['Quantity'] * data['Price']
data
```

```python
data['Date'] = pd.to_datetime(data['Date'])

data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year
data['Month_Name'] = data['Date'].dt.strftime('%B')  # e.g., January, February

data[['Date', 'Month', 'Year', 'Month_Name']].head()
```

```python

monthly_sales = data.groupby('Month_Name')['Total_Amount'].sum()


monthly_sales_sorted = monthly_sales.sort_values(ascending=False)

highest_month = monthly_sales_sorted.idxmax()
lowest_month = monthly_sales_sorted.idxmin()

print(" Highest Sales Month:", highest_month)
print(" Lowest Sales Month:", lowest_month)


print("\n Monthly Sales Summary:\n")
print(monthly_sales_sorted)
```

```python

category_sales = data.groupby('Category')['Total_Amount'].sum()

category_sales_sorted = category_sales.sort_values(ascending=False)

top_category = category_sales_sorted.idxmax()

print(" Top-Selling Product Category:", top_category)

# Optional: Display all categories and their total sales
print("\n Sales by Product Category:\n")
print(category_sales_sorted)
```

```python

category_payment = pd.crosstab(data['Category'], data['Payment_Method'])

print(" Category vs Payment Method:\n")
print(category_payment)
```

```python
repeat_customers = data['Customer_ID'].value_counts()
repeat_customers[repeat_customers > 1]
```

```python
data.groupby('Region')['Total_Amount'].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style="whitegrid")

# Total Sales by Region
region_sales = data.groupby('Region')['Total_Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
region_sales.plot(kind='bar', color='skyblue', title='Total Sales by Region')
plt.ylabel('Total Sales')
plt.xlabel('Region')
plt.tight_layout()
plt.show()

```

```python
# Top 10 Customers by Total Spending
top_customers = data.groupby('Customer_ID')['Total_Amount'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
top_customers.plot(kind='bar', color='green', title='Top 10 Customers by Spending')
plt.ylabel('Total Spending')
plt.xlabel('Customer ID')
plt.tight_layout()
plt.show()
```

```python
monthly_sales = data.groupby('Month_Name')['Total_Amount'].sum()
month_order = ['January', 'February', 'March', 'April', 'May', 'June']
monthly_sales = monthly_sales.reindex(month_order)
plt.figure(figsize=(10, 5))
monthly_sales.plot(kind='line', marker='o', color='purple', title='Monthly Sales Trend')
plt.ylabel('Total Sales')
plt.xlabel('Month')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
repeat_counts = data['Customer_ID'].value_counts()
repeaters = repeat_counts[repeat_counts > 1]
plt.figure(figsize=(8, 4))
repeaters.value_counts().sort_index().plot(kind='bar', color='orange', title='Repeat Customer Frequency')
plt.ylabel('Number of Customers')
plt.xlabel('Number of Purchases')
plt.tight_layout()
plt.show()
```

```python
category_payment = pd.crosstab(data['Category'], data['Payment_Method'])
category_payment.plot(kind='bar', stacked=True, figsize=(10, 6), title='Payment Methods by Product Category')
plt.ylabel('Number of Transactions')
plt.xlabel('Product Category')
plt.legend(title='Payment Method')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_cols = ['Quantity', 'Price', 'Total_Amount']

corr_matrix = data[numerical_cols].corr()


plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Correlation Heatmap: Quantity, Price, and Total_Amount")
plt.tight_layout()
plt.show()
```
